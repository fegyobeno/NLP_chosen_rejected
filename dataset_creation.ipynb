```python
system_prompt = """You are an expert dataset creator.
Task:
1. Analyze the provided [Input Text].
2. Create a specific Question relative to that text.
3. Construct a valid JSON object with:
   - "prompt": A list containing a user role with the formatted string "Forrás szöveg: {text} Kérdés: {question}"
   - "chosen": A list containing an assistant role with the correct answer derived from the text.
   - "rejected": A list containing an assistant role with an answer that is factually incorrect or contradicts the text.
"""
```

```python
def generate_chosen_rejected(data):
    try:

        chat_completion = client.chat.completions.create(
            model="zai-org/GLM-4.5-Air-FP8",
            messages=[
                {"role": "system", "content": system_prompt },
                {
                    "role": "user", 
                    "content": f"{few_shot_prompt}\n\n### New Task\n[Input Text]:\n\"{data}\"\n\n[Target JSON]:"
                }
            ],
            temperature=0.7,
            stream=False,
            response_format={"type": "json_object" }
        )
        answer_json = chat_completion.choices[0].message.content
    except Exception as e:
        print(f"Error generating dataset entry: {e}")
        answer_json = {}
    return answer_json
```

```python
def generate_chosen_rejected_dataset(input_file, output_file):
    with open(input_file, "r", encoding="utf-8") as f :
        data = json.load(f)
    
    dataset = {}

    for item in data.keys():
        #print(item, data[item])
        t = data[item]
        ans = generate_chosen_rejected(t)
        #print(f"Generated for item {item}: {ans}")
        try:
            json_ans = json.loads(ans)
            # Check if keys exist
            if not all(k in json_ans for k in ["prompt", "chosen", "rejected"]):
                raise ValueError("Missing keys")
            else:
                print(f"Successfully parsed generation for item {item}.")
                dataset[item] = json_ans
          
            
        except Exception as e:
            print(f"Failed to parse generation: {e}")

    with open(output_file, "w", encoding="utf-8") as f:
        json.dump(dataset, f, ensure_ascii=False, indent=4)
        
```
